```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(school_data, student_data, how="inner", on=["school_name"])
```

```python
# Calculate the budget total under school_data
budget_total = school_data["budget"].sum()

# Calculate the total schools under school_data
total_schools = len(school_data["school_name"].unique())

# Calculate the total number of Students under student_data_complete
grouped = school_data_complete.groupby("school_name")["budget"].apply(lambda x: x.count())
students = grouped.sum()

# Calculate the Average Math score under student_data
avg_math = student_data["math_score"].mean()

# Calculate the Average Reading score under student_data
avg_reading = student_data["reading_score"].mean()

# Calculate the Overall passing score with calculated scores
overall_passing = (avg_reading + avg_math)/2

# Create a Dataframe with a summary
district_summary_table = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [students],
    "Total Budget": [budget_total],
    "Average Math Score": [avg_math],
    "Average Reading Score": [avg_reading],
    "Overall Passing Rate": [overall_passing]
    })

district_summary_table
```

```python
top_per_group = school_data_complete.groupby("school_name")

top_per = top_per_group.agg({
    "type": "unique",
    "budget": "sum",
    "student_name": "count",
    "math_score": "mean",
    "reading_score": "mean" })

top_per["passing_reading_pct"] = top_per_group.apply(lambda x: (x["reading_score"] >= 70).sum() / x["student_name"].count()) * 100
top_per["passing_math_pct"] = top_per_group.apply(lambda x: (x["math_score"] >= 70).sum() / x["student_name"].count()) * 100

top_per["Overall Passing Rate"] = (top_per["passing_reading_pct"] + top_per["passing_math_pct"])/2
top_per
result_top = top_per.sort_values(by='Overall Passing Rate', ascending=False, na_position='first')

result_top.head()
```
