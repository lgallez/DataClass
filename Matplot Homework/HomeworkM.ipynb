```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

city_data = pd.read_csv("data/city_data.csv")
ride_data = pd.read_csv("data/ride_data.csv")

pyber_data = pd.merge(ride_data, city_data, how="left", on=["city", "city"])
pyber_data.head()
```

```python
urban = pyber_data[(pyber_data["type"]=="Urban")]
rural = pyber_data[(pyber_data["type"]=="Rural")]
suburban = pyber_data[(pyber_data["type"]=="Suburban")]

urban_ride_count = urban.groupby(["city"]).count()["ride_id"]
urban_fare = urban.groupby(["city"]).mean()["fare"]
urban_driver_count = urban.groupby(["city"]).count()["driver_count"]

suburban_ride_count = suburban.groupby(["city"]).count()["ride_id"]
suburban_fare = suburban.groupby(["city"]).mean()["fare"]
suburban_driver_count = suburban.groupby(["city"]).count()["driver_count"]

rural_ride_count = rural.groupby(["city"]).count()["ride_id"]
rural_fare = rural.groupby(["city"]).mean()["fare"]
rural_driver_count = rural.groupby(["city"]).count()["driver_count"]
```

```python
plt.scatter(urban_ride_count,
            urban_fare,
            s=10*urban_driver_count, c="gold",
            edgecolor="black", linewidths=1, marker="o",
            alpha=0.8, label="Urban")

plt.scatter(suburban_ride_count,
            suburban_fare,
            s=10*suburban_driver_count, c="lightskyblue",
            edgecolor="black", linewidths=1, marker="o",
            alpha=0.8, label="Suburban")

plt.scatter(rural_ride_count,
            rural_fare,
            s=10*rural_driver_count, c="coral",
            edgecolor="black", linewidths=1, marker="o",
            alpha=0.8, label="Rural")

plt.title("Pyber Ride Share Data (2016)")
plt.ylabel("Average Fare")
plt.xlabel("Total Riders per City")
plt.grid(True)

legend = plt.legend(fontsize="small", mode="Expanded",
                   numpoints=1, scatterpoints=1,
                   loc="best", title="City Types",
                   labelspacing=.75)

legend.legendHandles[0]._sizes = [75]
legend.legendHandles[1]._sizes = [75]
legend.legendHandles[2]._sizes = [75]

plt.text(42, 40, "Note:\n Circle size correlates with\n driver count per city.")

plt.savefig("scatterplot.png")
plt.show()
```

```python
fares_percent = pyber_data.groupby(["type"]).sum()["fare"] *100

plt.pie(fares_percent,
       labels=["Rural", "Suburban", "Urban"],
       colors=["gold", "lightskyblue", "coral"],
       explode=[0, 0, .1],
       autopct="%1.1f%%",
       shadow=True,
       startangle=150
       )
plt.title("% of Total Fares by City Type")

plt.savefig("faresbycity.png")
plt.show()
```

```python
rides_percent = pyber_data.groupby(["type"]).count()["ride_id"] / pyber_data["ride_id"].count() *100

plt.pie(rides_percent,
       labels=["Rural", "Suburban", "Urban"],
       colors=["gold", "lightskyblue", "coral"],
       explode=[0, 0, .1],
       autopct="%1.1f%%",
       shadow=True,
       startangle=150
       )
plt.title("% of Total Rides by City Type")

plt.savefig("ridesbycity.png")
plt.show()
```

```python
drivers_percent = city_data.groupby(["type"]).sum()["driver_count"] / city_data["driver_count"].sum() *100

plt.pie(drivers_percent,
       labels=["Rural", "Suburban", "Urban"],
       colors=["gold", "lightskyblue", "coral"],
       explode=[0, 0, .1],
       autopct="%1.1f%%",
       shadow=True,
       startangle=150,
       pctdistance=.7
       )
plt.title("% of Total Drivers by City Type")

plt.savefig("driversbycity.png")
plt.show()
```
